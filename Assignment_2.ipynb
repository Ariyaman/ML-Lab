```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, plot_confusion_matrix, plot_roc_curve
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from itertools import cycle
from sklearn.preprocessing import label_binarize
```

Iris Dataset

```python
# Dataset Preparation
iris_dataset = datasets.load_iris()

print("Iris Dataset Feature and target names - \n")
print(iris_dataset.feature_names)
print(iris_dataset.target_names)

X = iris_dataset.data
Y = iris_dataset.target

X_train_1,X_test_1,Y_train_1,Y_test_1 = train_test_split(X,Y,test_size=0.3) # 70:30 
X_train_2,X_test_2,Y_train_2,Y_test_2 = train_test_split(X,Y,test_size=0.4) # 60:40
X_train_3,X_test_3,Y_train_3,Y_test_3 = train_test_split(X,Y,test_size=0.5) # 50:50
X_train_4,X_test_4,Y_train_4,Y_test_4 = train_test_split(X,Y,test_size=0.6) # 40:60
X_train_5,X_test_5,Y_train_5,Y_test_5 = train_test_split(X,Y,test_size=0.7) # 30:70

pca = PCA(n_components=3)

X_with_PCA = pca.fit_transform(X)

X_train_pca_1,X_test_pca_1,Y_train_pca_1,Y_test_pca_1 = train_test_split(X_with_PCA,Y,test_size=0.3) # 70:30 
X_train_pca_2,X_test_pca_2,Y_train_pca_2,Y_test_pca_2 = train_test_split(X_with_PCA,Y,test_size=0.4) # 60:40
X_train_pca_3,X_test_pca_3,Y_train_pca_3,Y_test_pca_3 = train_test_split(X_with_PCA,Y,test_size=0.5) # 50:50
X_train_pca_4,X_test_pca_4,Y_train_pca_4,Y_test_pca_4 = train_test_split(X_with_PCA,Y,test_size=0.6) # 40:60
X_train_pca_5,X_test_pca_5,Y_train_pca_5,Y_test_pca_5 = train_test_split(X_with_PCA,Y,test_size=0.7) # 30:70
```
